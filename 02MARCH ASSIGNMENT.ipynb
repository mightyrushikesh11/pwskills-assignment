Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans. Matplotlib is a Python library used for data visualization. It provides a wide range of tools for creating visualizations such as line plots, scatter plots, bar plots, histograms, and many more. Matplotlib is a highly customizable and versatile library that allows users to create high-quality plots suitable for scientific publications, reports, and presentations.

Matplotlib is used to visualize data to gain insights into the underlying patterns and relationships in the data. It is commonly used in scientific research, data analysis, and machine learning applications. With Matplotlib, users can create highly customizable plots that are easy to understand and interpret.

The following are five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot: A line plot is a graph that displays data as a series of points connected by a line. It is used to display trends over time or to compare two or more variables.

Scatter Plot: A scatter plot is a graph that displays data as a collection of points. It is used to visualize the relationship between two variables.

Bar Plot: A bar plot is a graph that displays data as a series of bars. It is used to compare the values of different categories.

Histogram: A histogram is a graph that displays the distribution of a dataset. It is used to visualize the frequency of data in different bins.

Pie Chart: A pie chart is a circular graph that displays data as a series of slices. It is used to visualize the composition of a dataset.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a graph that displays data as a collection of points. It is used to visualize the relationship between two variables.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('SCATTER PLOT')
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create multiple plots in the same figure. It allows us to divide a figure into a grid of subplots and plot different data in each subplot. This is useful for comparing multiple datasets or visualizing different aspects of a single dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(nrows=2, ncols=2)

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

fig.suptitle('Line Plots')

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Bar Plot: A bar plot is a graph that displays data as a series of bars. It is used to compare the values of different categories.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.xlabel('company')
plt.ylabel('profit')
plt.bar(company,profit)
plt.show()
plt.xlabel('company')
plt.ylabel('profit')
plt.barh(company,profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that displays its five-number summary, which includes the minimum and maximum values, the first quartile (Q1), the median (Q2), and the third quartile (Q3). The plot also shows any outliers in the dataset, which are values that are significantly different from the rest of the data.

Box plots are used to summarize the distribution of a dataset and to compare the distributions of multiple datasets. They are particularly useful when dealing with large datasets or when comparing data across different groups or categories.

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
plt.xlabel('X- axis')
plt.ylabel('Y-axis')
plt.title('BOX PLOT')
plt.boxplot(data)
plt.show()
```
