Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans.Flask is a micro web framework written in Python that allows developers to quickly and easily build web applications. It is designed to be simple, flexible, and lightweight, making it an ideal choice for building small to medium-sized web applications.

Here are some advantages of using the Flask framework:

Lightweight and Flexible: Flask is a lightweight and flexible framework that allows developers to build web applications quickly and easily. It has a small and simple core, with optional extensions that can be added to provide additional functionality.

Easy to Learn: Flask is one of the easiest frameworks to learn, thanks to its simplicity and small codebase. It is an excellent choice for developers who are new to web development or those who want to get started quickly.

Extensible: Flask is extensible, which means that developers can add their own functionality to the framework easily. There are many extensions available for Flask that can be used to add features like authentication, database integration, and more.

Pythonic: Flask is built on Python, which means that it follows the same design principles as the Python language. This makes it easy for Python developers to learn and use Flask.

Easy to Test: Flask provides an excellent testing framework that makes it easy to write unit tests and integration tests for web applications. This allows developers to catch bugs early and ensure that their applications are functioning as intended.

Well Documented: Flask has excellent documentation that makes it easy for developers to get started with the framework. The documentation covers everything from basic usage to advanced topics like extension development.

Q3. What is App routing in Flask? Why do we use app routes?

Ans.In Flask, app routing refers to the process of associating URL patterns with functions that will be executed when a user visits a specific URL. This is done using the @app.route decorator in Flask. The URL pattern is specified as an argument to the decorator, and the function that should be executed when the URL is visited is defined directly below the decorator.

We use app routes in Flask to create a logical structure for our application, making it easy for users to navigate between different pages and for developers to manage the application code. By using app routes, we can create a modular design that separates different components of the application and makes it easier to maintain and update the code. Additionally, app routing makes it easy to handle complex URLs with parameters and query strings, and allows us to create RESTful web services that can be consumed by other applications.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the 
url_for() function.

In Flask, the url_for() function is used for URL building. It generates a URL for a given view function based on the name of the function.

Here is an example Python code that demonstrates the use of the url_for() function in Flask:


from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the home page!'

@app.route('/about')
def about():
    return 'This is the about page!'

@app.route('/user/<username>')
def user_profile(username):
    return f'This is {username}\'s profile page'

with app.test_request_context():
    print(url_for('home')) # prints '/'
    print(url_for('about')) # prints '/about'
    print(url_for('user_profile', username='john')) # prints '/user/john'

    
In this code, we first import the url_for() function from the Flask module. We then define three different routes using the @app.route decorator: one for the home page (/), one for the about page (/about), and one for user profiles (/user/<username>). The user profile route takes a dynamic URL parameter, username.

We then create a test request context using the test_request_context() method, which allows us to test the url_for() function without actually making a request to the application.

Finally, we call the url_for() function three times, passing in the name of the view function and any necessary parameters. The function returns a string representing the URL for that view function, which we print to the console.
