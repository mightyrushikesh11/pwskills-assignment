Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Ans.In Bokeh, glyphs are graphical marks used to represent data visually in a plot. Examples of glyphs include circles, squares, lines, bars, and more. Bokeh provides a wide range of glyph types to choose from, which can be customized by specifying various attributes like size, color, and alpha.

```python
# I am suing circle glyphs to plot a graph.

import bokeh.io
import bokeh.plotting
from bokeh.plotting import output_file,show,figure
bokeh.io.output_notebook()

output_file('Graph.html')

p= figure(title='SCATTER PLOT')
p.xaxis.axis_label='X-axis'
p.yaxis.axis_label='Y-axis'
x=[1,2,3,4,5]
y=[6,7,8,9,10]

p.circle(x=x,y=y)

show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in 
real time?


Bokeh server is a server-side library that allows you to create and deploy interactive data visualizations on the web. It allows you to create interactive plots that can be updated in real-time by connecting to data sources, handling user inputs, and updating the plot dynamically. Bokeh server provides the ability to create reactive, real-time data applications that can handle user interactions and update the plot accordingly.

To create interactive plots using Bokeh server, you can follow these steps:

Define a plot with a data source
Create a function that updates the plot when triggered by user input
Create a Bokeh application that defines the layout of the plot and the input widgets
Run the Bokeh application using the Bokeh server
Here's an example of how to create an interactive plot using Bokeh server:

```python
from bokeh.io import curdoc
from bokeh.layouts import row, column
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure,show
from bokeh.models.widgets import Slider

# Define the data source
source = ColumnDataSource(data=dict(x=[1, 2, 3], y=[1, 4, 9]))

# Define the plot
plot = figure(plot_height=400, plot_width=400, title="Plot Title")
plot.line('x', 'y', source=source, line_width=3)

# Define the update function
def update(attr, old, new):
    # Get the new slider value
    a = slider.value
    # Update the data source with the new values
    source.data = dict(x=[1, 2, 3], y=[a, 2*a, 3*a])

# Define the input widget
slider = Slider(start=0, end=10, value=1, step=1, title="Slider Title")
slider.on_change('value', update)

# Define the layout of the plot and the input widget
layout = column(slider, plot)

# Define the Bokeh application
curdoc().add_root(layout)

```

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?


Bokeh plots can be easily embedded into web pages or dashboards using Flask or Django frameworks. Here are the general steps to embed a Bokeh plot in a web page or dashboard:

Create a Bokeh plot as usual using Bokeh's plotting functions.

Define a Flask or Django route that returns the HTML code needed to display the plot in a web page or dashboard.

In the Flask or Django template where the plot should be displayed, include a placeholder element, such as a div tag, with a unique ID to identify the location where the plot should be inserted.

Use JavaScript code to initialize the Bokeh plot and attach it to the placeholder element identified by its ID. This code can be included in the same Flask or Django template where the placeholder element is defined, or in a separate JavaScript file.
