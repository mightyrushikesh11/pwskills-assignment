Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans. Five plots of Seaborn library is given below:

Scatter plot: A scatter plot is a two-dimensional plot that displays the relationship between two continuous variables. It is used to identify patterns and relationships between variables and to detect outliers.

Line plot: A line plot is a type of chart that displays information as a series of data points connected by straight lines. It is often used to track changes over time and to compare trends across multiple variables.

Bar plot: A bar plot is a chart that represents data with rectangular bars. It is used to compare the values of different categories or to show changes in data over time.

Histogram: A histogram is a graphical representation of the distribution of a dataset. It is used to visualize the shape of the distribution, the central tendency, and the spread of the data.

Heatmap: A heatmap is a graphical representation of data in which the values are represented by colors. It is used to identify patterns in data and to visualize correlations between variables. Heatmaps are often used in exploratory data analysis and in machine learning applications.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

fmri=sns.load_dataset('fmri')
fmri
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='region')
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
sns.boxplot(x='pclass',y='age',data=titanic)
```

```python
sns.boxplot(x='pclass',y='fare',data=titanic)
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamonds=sns.load_dataset('diamonds')
diamonds
```

```python
sns.histplot(data=diamonds,x='price',hue='cut',multiple='stack')
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
iris=sns.load_dataset('iris')
iris
```

```python
sns.pairplot(data=iris,hue='species')
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights=sns.load_dataset('flights')
flights
```

```python
flight=flights.pivot("month", "year", "passengers")
flight
```

```python
sns.heatmap(data=flight)
```
